```python
import os
from moviepy.editor import VideoFileClip
import math

def segment_videos(input_dir: str, output_dir: str, clip_duration_seconds: int):
    print(f"Starting video segmentation...")
    print(f"Input Directory: {input_dir}")
    print(f"Output Directory: {output_dir}")
    print(f"Clip Duration: {clip_duration_seconds} seconds")
    print("-" * 30)

    os.makedirs(output_dir, exist_ok=True)
    supported_extensions = ('.mp4', '.avi', '.mov', '.mkv', '.flv')
    video_files = [f for f in os.listdir(input_dir) if f.lower().endswith(supported_extensions)]

    if not video_files:
        print("No video files found in the input directory. Please check the path.")
        return

    for filename in video_files:
        video_path = os.path.join(input_dir, filename)
        try:
            with VideoFileClip(video_path) as video:
                video_duration = video.duration
                video_name = os.path.splitext(filename)[0]
                print(f"\nProcessing '{filename}' (Duration: {video_duration:.2f}s)")
                num_clips = math.floor(video_duration / clip_duration_seconds)
                if num_clips == 0:
                    print(f"  -> Skipping '{filename}', as it's shorter than the clip duration.")
                    continue
                for i in range(num_clips):
                    start_time = i * clip_duration_seconds
                    end_time = start_time + clip_duration_seconds
                    output_filename = f"{video_name}_clip_{i+1:03d}.mp4"
                    output_path = os.path.join(output_dir, output_filename)
                    print(f"  -> Creating clip {i+1}/{num_clips}: {output_filename}")
                    new_clip = video.subclip(start_time, end_time)
                    new_clip.write_videofile(output_path, codec="libx264", audio_codec="aac", logger=None)
        except Exception as e:
            print(f"!!!!!! Error processing file {filename}: {e} !!!!!!")
            continue

    print("\n" + "=" * 30)
    print("✅ Video segmentation complete!")
    print(f"All clips have been saved to: {output_dir}")
    print("=" * 30)

INPUT_VIDEO_DIR = '/kaggle/input/video-based-classroom-attention-tracker/ocat dataset2'
OUTPUT_CLIPS_DIR = '/kaggle/working/segmented_clips2'
CLIP_DURATION = 3

if not os.path.exists(INPUT_VIDEO_DIR):
    print("Creating dummy input directory for demonstration purposes.")
    dummy_video_folder = os.path.join('/kaggle/working', 'dummy_videos')
    os.makedirs(dummy_video_folder, exist_ok=True)
    INPUT_VIDEO_DIR = '/kaggle/input/video-based-classroom-attention-tracker/ocat dataset2'
    print(f"INFO: Running with dummy path '{INPUT_VIDEO_DIR}'. Upload your data and change this path for real use.")

segment_videos(
    input_dir=INPUT_VIDEO_DIR,
    output_dir=OUTPUT_CLIPS_DIR,
    clip_duration_seconds=CLIP_DURATION
)
```
